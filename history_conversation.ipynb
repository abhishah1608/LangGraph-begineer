Conversation class of langchain memory.

```python
from langchain.memory import ConversationBufferMemory
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
conversation_memory = ConversationBufferMemory()
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.0)
conversation = ConversationChain(llm=llm, memory=conversation_memory, verbose=True)
```

```python
input_query = "Hello My name is Abhi shah."

conversation.predict(input=input_query)
```

```python
input_query = "Hello can you help me with job hunting? I am a senior Software Engineer with 8.5 years of experience."

conversation.predict(input=input_query)
```

```python
input_query = "Hello How are you feeling today."

conversation.predict(input=input_query)
```
