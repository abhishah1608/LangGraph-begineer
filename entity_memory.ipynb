```python
from langchain.chains import ConversationChain
from langchain.chat_models import ChatOpenAI    
from langchain.memory import ConversationEntityMemory
from dotenv import load_dotenv
from langchain.prompts import PromptTemplate
```

```python
load_dotenv()
```

```python
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.1)
```

```python
entity_memory = ConversationEntityMemory(llm=llm, k=3)
```

```python
entity_memory.load_memory_variables({'inputs': "Hi, My name is Abhi shah, How are you doing today?"})
```

```python
ENTITY_MEMORY_CONVERSATION_TEMPLATE = """
You are a helpful assistant.

Relevant entities:
{entities}

Current conversation:
{history}
Human: {input}
AI:
"""

prompt = PromptTemplate(
    input_variables=["entities", "history", "input"],
    template=ENTITY_MEMORY_CONVERSATION_TEMPLATE,
)

conversation = ConversationChain(
    llm=llm,
    prompt=prompt,
    memory=entity_memory,
    verbose=True
)
```

```python
conversation.predict(input="My name is Abhi. I am a software developer. I live in Montreal, Quebec")
```

```python
conversation.predict(input="Abhi is living in which city?")
```
